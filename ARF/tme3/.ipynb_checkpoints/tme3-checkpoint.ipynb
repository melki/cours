### Algorithme de descente de gradient

```python
from collections import namedtuple
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
Fonction = namedtuple("Fonction",["f","grad","dim"])
def f_id(x): return x
def f_id_grad(x): return 1
identite = Fonction(f_id,f_id_grad,1)
identite.f(1), identite.grad(1), identite.dim

def v2m(x):
    return x.reshape((1,x.size)) if len(x.shape)==1 else x
```

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def optimize(f,xinit=None,eps=0.01,max_iter=100):
    logX, logF, logGrad = np.zeros((max_iter,f.dim)),np.zeros((max_iter)),np.zeros((max_iter,f.dim))
    if(xinit == None):
        xinit  = np.random.random((f.dim))
    logX[0] = np.array(xinit)
    for i in range(1,max_iter):
        logGrad[i]=f.grad(logX[i-1])
        logX[i]=logX[i-1]-eps*logGrad[i]
        logF[i]=f.f(logX[i])
    return logX, logF, logGrad
```

```python
def x_Cosx(x):
    return (x * np.cos(x)).T
def x_Cosx_grad(x):
    return (np.cos(x) + x*(-np.sin(x))).T
xCosx = Fonction(x_Cosx,xCosx_grad,1)

def banana(x):
    return 100*(x[1]-x[0])**2 + (1-x[0])**2
def banana_grad(x):
    return [100*(-2*x[1] + 2*x[0]) + (-2+2*x[0]),100*(2*x[1] - 2*x[0])]
fban = Fonction(banana,banana_grad,2)
```

```python
a,b,c = optimize(xCosx,xinit=17.,eps=.1)
x = np.linspace(-10.,20,100)
xx = np.linspace(5.,20,100)
plt.plot(x,xCosx.f(x))
plt.plot(a[-1],b[-1],'o')
plt.show()
```

```python
a,b,c = optimize(fban,xinit=[-5,-5],eps=.1,max_iter=30)
print a[-1]
fig = plt.figure()
ax = fig.gca(projection='3d')
x = np.linspace(-10.,20,100)
ax.plot(x, x, fban.f([x,x]))
plt.plot([1],[6,1],'o')

plt.show()
```
