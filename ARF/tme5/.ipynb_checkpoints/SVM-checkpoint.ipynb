# String Kernel


```python
def getBigram(t):
    ngram = []
    for tt in t:
        for j in range(0,len(tt)):
            for k in range(j+1,len(tt)):
                if tt[j]+tt[k] not in ngram:
                    ngram.append(tt[j]+tt[k])
    return ngram
    
def getDict(t):
    d=[]
    for j in range(0,len(t)):
            for k in range(j+1,len(t)):
                d.append((t[j]+t[k],k-j))
                
    return d

def getSim(t1,t2,l):
    
    d1 = getDict(t1)
    d2 = getDict(t2)
    
    
    
    sim = 0
    for a in d2:
        for b in d1:
    
            if a[0]==b[0]:
                w.append(a[0])
                sim += l ** (a[1]+b[1])
        
    return sim,w
```
