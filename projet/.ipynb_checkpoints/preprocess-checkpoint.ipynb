```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data.csv", sep=';')
df
```

```python
an = []
for i in df["date du scrutin"]:
    j = str(i)
    j = j.split('-')
    if(j[0]!='nan'):
        an.append(+int(j[0]))
    else:
        an.append("nan")
df["annee"] = an
```

```python
typel = []
for i in df["libelle du scrutin"]:
    if(i!='nan'):
    
        j = str(i)
        j = j.split(' ')
        if(j[0]!='nan'):
            typel.append(j[0])
        else:
            typel.append("nan")
    else:
        typel.append("nan")
df["type"] = typel
```
